# GroupBy Operations

```python
df['Quarter'] = ['Q1', 'Q1', 'Q1', 'Q2', 'Q2','Q2', 'Q3', 'Q3', 'Q3']
```

```python
print(df)
```

```python
groups = df.groupby('Quarter')
```

```python
groups.size()
```

```python
print(groups.mean())
print(groups.max())
```

```python
print(groups.aggregate([min, max]).round(2))
```

```python
df['Odd_Even'] = ['Odd', 'Even', 'Odd', 'Even', 'Odd', 'Even','Odd', 'Even', 'Odd']
groups = df.groupby(['Quarter', 'Odd_Even'])
```

```python
print(df)
```

```python
groups = df.groupby(['Quarter', 'Odd_Even'])
print(groups)
```

```python
print(groups[['col1', 'col4']].aggregate([sum, np.mean]))
```

# Concatenation, Joining, and Merging

```python
df1 = pd.DataFrame(['100', '200', '300', '400'],index=['a', 'b', 'c', 'd'],columns=['A',])
df2 = pd.DataFrame(['200', '150', '50'],index=['f','d','b'],columns=['B',])
print(df1)
print(df2)
```

```python
print(df1.append(df2, sort=False))
```

```python
print(pd.concat((df1, df2), sort=False))
```

```python
print(df1.join(df2))
print(df2.join(df1))
```

```python
print(df1.join(df2, how='left'))
print(df1.join(df2, how='right'))
print(df1.join(df2, how='inner'))
print(df1.join(df2, how='outer'))
```

# Merging

```python
c = pd.Series([250, 150, 50], index=['b', 'd', 'c'])
df1['C'] = c
df2['C'] = c
```

```python
print(df1)
print(df2)
```

```python
print(pd.merge(df1, df2, how='outer'))
print(pd.merge(df1, df2))
print(pd.merge(df1, df2, on='C'))
```

```python
print(pd.merge(df1, df2, left_on='A', right_on='B'))
print(pd.merge(df1, df2, left_on='A', right_on='B', how='outer'))
```
